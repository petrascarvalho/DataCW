```python
# Importando as bibliotecas para o projeto

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Realizando a leitura do dataset 'titanic_dataset.csv' disponível no ambiente Colab

df = pd.read_csv('/content/Titanic/titanic_dataset.csv')
```

```python
# Exibindo as primeiras linhas do dataset para verificação

print(display(df.head()))
```

```python
# Buscando valores duplicados

valores_duplicados = df[df.duplicated()]
print(len(valores_duplicados))
```

```python
# Buscando valores nulos

valores_nulos = df[df.isnull().any(axis=1)]
print(valores_nulos)
```

```python
# substituindo NaN por 0

df.replace(np.nan, 0, inplace=True)
print(df)
```

```python
# Removendo campos nulos

df.dropna(inplace=True)
print(df)
```

```python
# Exibindo dataset para verificação

print(display(df.head()))
```

```python
# Ordenando o dataset por Genero

sorted_df = df.sort_values(by='Sex', ascending=True)
print(sorted_df)
```

```python
# Agrupando o dataset por idade

grouped_by = df.groupby('Age')
print(grouped_by.head())
```

```python
# Exibindo a quantidade de sobreviventes

quantidade_sobreviventes = df['Survived'].value_counts()
print(quantidade_sobreviventes)


# Exibindo a quantidade de sobreviventes por idade
idade = df['Age'].value_counts()
print(idade)

# Exibindo a quantidade de sobreviventes por genero
genero = df['Sex'].value_counts()
print(genero)

# agrupando os sobreviventes por idade e genero
sobreviventes_por_idade = df.groupby(['Age', 'Sex'])['Survived'].value_counts()
print(sobreviventes_por_idade)

# Gerando um grafico com a relação entre idade e a quantidade de sobreviventes

survivors_by_age_counts = df[df['Survived'] == 1]['Age'].value_counts().sort_index()
plt.figure(figsize=(15, 7))
plt.bar(survivors_by_age_counts.index, survivors_by_age_counts.values)
plt.xlabel('Idade')
plt.ylabel('Quantidade de Sobreviventes')
plt.title('Quantidade de Sobreviventes por Idade')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Gerando um grafico com a relação entre genero e a quantidade de sobreviventes

survivors_by_age_counts = df[df['Survived'] == 1]['Sex'].value_counts().sort_index()
plt.figure(figsize=(15, 7))
plt.bar(survivors_by_age_counts.index, survivors_by_age_counts.values)
plt.xlabel('Genero')
plt.ylabel('Quantidade de Sobreviventes')
plt.title('Quantidade de Sobreviventes por Genero')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Realizando a remoção de outliers por idade

sns.boxplot(x=df['Age'])
plt.show()
```

```python
# Apos a analise, a quantidade de sobrevivente com 0 anos, foi muito discrepante em relação as outras idades.
# Exibindo a quantidade de sobreviventes por idade menor que 1
idade = df[df['Age'] < 1]['Age'].value_counts()
print(idade)
```
